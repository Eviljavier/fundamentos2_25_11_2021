#Estructura de seleccion de bifurcacion (if-else)

Esta estructura,selecciona una de dos opcioes de ejecuciones, se ejecuta un bloque si la condicion se cumple o se ejecuta otro bloque si la condicion no se cumple(else)

```
if<condicion> :
    <bloque de ejecucuion>
else:    
    <bloque de ejecucuion alterno>



````python
edad = int(input("Dame tu edad"))
print(f"Tecleaste {edad} años")
if edad >= 18:
    print("Eres mayor de edad, ten tu cheve")
else:
    print("Eres menor de edad, ten tu boing de mango")    
print("Fin del programa")
````
